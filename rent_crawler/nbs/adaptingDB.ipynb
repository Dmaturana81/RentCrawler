```python
not_arrays = [x for x in prices if not isinstance(x['prices'],list)]
```

```python
len(not_arrays)
```

```python
not_arrays[0]
```

```python
def fixing_prices(data):
    """Function that transform a house document into prices document"""
    return {
        # '_id':data['_id'],
        'code':data['code'],
        'prices':[data['prices']]
    }
```

```python
fixing_prices(not_arrays[0])
```

```python
db['houses_prices'].replace_one({'_id': ObjectId(not_arrays[0]['_id'])}, fixing_prices(not_arrays[0]))
```

```python
list(db['houses_prices'].find({'code':'QA_894446935'}))
```

```python
for prop in tqdm(not_arrays):
    db['houses_prices'].replace_one({'_id': ObjectId(prop['_id'])}, fixing_prices(prop))
```
